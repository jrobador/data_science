```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Random data
np.random.seed(21)

#Define number of features (X) and number of outputs (Y)
X = 2 * np.random.rand(50, 1)
y = 4 + 3 * X + np.random.randn(50, 1)

X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)

print (X.shape[0], y.shape[1])
print (X)
```

```python
#Linear regression model
#Given the linear equation: Y = W x = B
#Inputs of the model = y and x.
#Ouputs of the model = W and B.

class LinearRegression(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearRegression, self).__init__()
        self.linear = nn.Linear(input_dim, output_dim)
        
    def forward(self, x):
        return self.linear(x)

model = LinearRegression(X.shape[1], y.shape[1])
```

```python
#Loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
num_epochs = 10000
for epoch in range(num_epochs):
    # Forward pass
    outputs = model(X_tensor)
    loss = criterion(outputs, y_tensor)
    
    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
#Parameters of model
w, b = model.parameters()

print("Weight:", w.item())
print("Bias:", b.item())

w = w.detach().numpy()
b = b.detach().numpy()
```

```python
# Plots
plt.scatter(X, y)
plt.plot(X, w*X + b, color='red')  # Regression line
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```
